```python
import numpy as np
import matplotlib.pyplot as plt

wavelen, sol = np.loadtxt ("cosi_full_1nm_max.txt", unpack=True)

plt.plot(wavelen, np.log(sol), color='red', label='solar spectrum', linewidth='0.5', marker='o', markersize='1.5')
plt.plot(wavelen, np.log(sol**0.5), color='blue', label='solar spectrum', linewidth='1.5')
plt.yscale = 'log'
plt.legend()
plt.show()
```
